```python
import requests
import json
import pandas as pd
import numpy as np
import math
```

```python
path = '.\customers.txt'
```

```python
with open(path, 'rb') as f:
    data = f.readlines()
    
data = map(lambda x: x.rstrip(), data)
```

```python
data_json_str = "[" + ','.join(data) + "]"
```

```python
customers = pd.read_json(data_json_str)
```

```python
customers = customers.sort_values(by = 'user_id', ascending = True).reset_index(drop = True)
```

```python
customers
```

```python
def dist_function(loc):
    names = []
    uids = []
    
    for i in range(len(customers)):
        degrees_to_radians = math.pi/180.0
        
        phi1 = (90 - loc[0])*degrees_to_radians
        phi2 = (90 - customers.ix[i][0]) * degrees_to_radians
        
        name = customers.ix[i][2]
        uid = customers.ix[i][3]
        
        
        lambda1 = loc[1]*degrees_to_radians
        lambda2 = customers.ix[i][1]*degrees_to_radians
        lambda_diff = lambda1 - lambda2
        
        central_angle = np.arccos(np.sin(phi1)*np.sin(phi2) + np.cos(phi1)*np.cos(phi2)*np.cos(lambda_diff))
        
        distance = 6371 * central_angle

        if distance <= 100:
            names.append(name)
            uids.append(uid)
            
    within_100_df = pd.DataFrame({'name': names, 'user_id': uids}, columns = ['name', 'user_id'])
    print within_100_df
```

The table would consist of rows corresponding to a user (most likely indicated by user_id) and columns for which tier they are in and for each possible product. The values for the "tier" column would simply consist of "lite", "standard", or "premium", whereas the values for the "product" columns would be binary values (either 0 or 1) depending on whether or not the customer has purchased the product. There should also be a column populated by datetime values, pertaining to the specific time and date that customer made a purchase, and consequently when the row was created. Thus, you can group by individual users to see if there are any changes in the tiers and products they've purchased. 

As implied earlier, the table would get populated each time a user made a purchase. When this population happens, the table would be given the following:

- a unique id for the user for the "user_id" column
- a value for the "tier" column ("lite", "std", or "prem")
- a 0 or 1 for each product column ("learn", "engage", "support") depending on if the customer purchased said product
- a datetime value corresponding to the date and time the row is inserted into the table

This would occur every time a purchase was made, even if the same user has made multiple purchases. This is useful as we could see the changes, or lack thereof, the user has made over time. Of course, we can also group by the "tier" column and aggregate the users to see how many users fall in each tier category, and since we have a datetime column we can sort this by date too.
